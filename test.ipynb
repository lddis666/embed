```python
import bz2
import csv
from collections import Counter

DATA_PATH = ''  # 如果在当前目录就留空，否则写成 '/path/to/dir/'
FILENAME = '/mlx_devbox/users/liurundong.991/playground/embed/20250901.as-rel2.txt.bz2'

rel_values = []

with bz2.open(DATA_PATH + FILENAME, mode='rt') as f:  # 'rt' = 文本模式
    reader = csv.reader(f, delimiter='|')
    for row in reader:
        # 跳过注释行
        if not row or row[0].startswith('#'):
            continue
        # 有些行可能不规范，先防御性判断长度
        if len(row) < 3:
            continue
        try:
            rel = int(row[2])
            rel_values.append(rel)
        except ValueError:
            # 如果 row[2] 不是整数，跳过
            continue

# 统计不同关系值
counter = Counter(rel_values)
print("不同关系类型及其数量：")
for rel_type, count in sorted(counter.items()):
    print(f"关系值 {rel_type}: {count} 条")

print("\n总记录数：", len(rel_values))
print("出现过的关系值集合：", sorted(counter.keys()))
```

```python
import pandas as pd

def count_ones_in_column(file_path, column_name):
    """
    统计 CSV 文件中，指定列的值为 1 的行数。

    :param file_path: CSV 文件路径
    :param column_name: 需要统计的列名（字符串）
    :return: 值为 1 的行的数量（int）
    """
    df = pd.read_csv(file_path)
    # 注意：如果该列读进来是字符串类型，可以先转成数字
    # df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    return (df[column_name] == 1).sum()
```
